### Cleaning Data

Pertama, kita hapus tabel yang tidak diperlukan. Karena perbedaan hour_df dan day_df hanya pada kolom hr, dan tidak diperlukan untuk menjawab pertanyaan, maka bisa dihapuskan saja

```python
del hour_df
```

Setelah itu, hapus kolom yang tidak diperlukan pada day_df

```python
drop_col = ['instant', 'windspeed']

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

Untuk mempermudah pembacaan nama kolom, maka nama-nama tersebut bisa diubah agar lebih mudah dipahami

```python
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)

day_df.head()
```

Selain itu, data yang bersifat numerik juga bisa diubah menjadi kategorikal

```python
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Few clouds/Partly Cloudy',
    2: 'Misty/Cloudy/Broken clouds',
    3: 'Light Snow/Light Rain/Scattered clouds',
    4: 'Heavy Rain/Thunderstorm/Snow/Fog'
})
```

```python
day_df['dateday'] = pd.to_datetime(day_df.dateday)

day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_cond'] = day_df.weather_cond.astype('category')

day_df.head()
```

Lalu, cek kembali tipe data yang sudah diganti

```python
day_df.info()
```

## Exploratory Data Analysis (EDA)

Sekarang, kelompokkan semua pengguna sepeda berdasarkan bulan

```python
day_df.groupby(by='month').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

Berdasarkan data diatas, dapat ditarik kesimpulan bahwa rata-rata pengguna sepeda paling banyak ada di bulan Juni dan yang paling sedikit di bulan Januari

Kelompokkan penggunaan sepeda berdasarkan cuaca

```python
day_df.groupby(by='weather_cond').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

Dapat dilihat bahwa sepeda lebih banyak disewakan pada saat cuaca cerah

Selanjutnya, kelompokkan berdasarkan holiday

```python
day_df.groupby(by='holiday').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

Dapat dilihat bahwa penyewaan sepeda lebih banyak pada hari biasa dibandingkan pada hari libur

Lalu, bandingkan pada saat weekday dan weekend

```python
day_df.groupby(by='weekday').agg({
    'count':['max','min','mean']
})
```

Dari data diatas, dapat dilihat jika jumlah penyewa terbesar ada di hari Jumat

Kelompokkan berdasarkan workingday

```python
day_df.groupby(by='workingday').agg({
    'count': ['max', 'min', 'mean']
})
```

Jumlah penggunaan sepeda pada workingday dan non-workingday tidak berbeda jauh. Namun, penggunaan terbesar tetap pada workingday

Kelompokkan berdasarkan season

```python
day_df.groupby(by='season').agg({
    'casual': 'mean',
    'registered': 'mean',
    'count': ['max', 'min', 'mean']
})
```

Penggunaan terbanyak ada pada musim gugur (fall) dan yang tersedikit ada pada musim semi (spring)

Lalu, kelompokkan temp, atemp, dan hum berdasarkan season

```python
day_df.groupby(by='season').agg({
    'temp': ['max', 'min', 'mean'],
    'atemp': ['max', 'min', 'mean'],
    'hum': ['max', 'min', 'mean']
})
```

Dari hasil analisis diatas, dapat dilihat bahwa temp dan atemp yang paling tinggi ada pada musim gugur (fall), sedangkan humidity tertinggi ada pada musim dingin (winter)

Terakhir, untuk melihat hubungan antara penyewa yang casual, registered, beserta jumlahnya, dapat menggunakan kode dibawah ini

```python
fig, ax = plt.subplots(figsize=(10,6))
correlation_matrix = day_df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

sns.heatmap(
    correlation_matrix,
    annot=True,
    mask=mask,
    cmap="coolwarm",
    center=0,
    fmt=".3f")
plt.title("Correlation Heatmap")
plt.show()
```

Dari hasil analisis yang didapatkan, kesimpulan yang kita dapatkan yaitu:
- temp dan atemp sangat berkolerasi (0.992)
- kolerasi hum dengan temp dan atemp lemah, hanya 0,127 dan 0,140
- casual cukup berkolerasi dengan temp dan atemp (0.543 dan 0.544), namun bernilai negatif untuk hum (-0.077)
- registered juga cukup berkolerasi dengan temp dan atemp (0.540 dan 0.544), bernilai negatif untuk hum (-0.091) dan berkolerasi dengan casual (0.395)
- count berkolerasi kuat dengan temp, atemp, casual dan registered (0.627, 0.631, 0,673, dan 0.946), hanya saja untuk korelasi dengan hum bernilai negatif (-0.101)

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah faktor cuaca mempengaruhi jumlah penyewaan sepeda?

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='weather_cond',
    y='count',
    data=day_df,
    palette='Set3')

plt.title('Jumlah Penyewa Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Penyewa Sepeda')
plt.show()
```

Berdasarkan visualisasi diatas, dapat disimpulkan bahwa faktor cuaca mempengaruhi jumlah penyewaan sepeda. Pada cuaca cerah, penyewaan sepeda sangat laris, dan berkurang ketika cuaca sedang berkabut, dan ketika cuaca sedikit hujan/bersalju, penyewaan sepeda menurun secara signifikan

### Pertanyaan 2: Bagaimana faktor temp, atemp, dan hum mempengaruhi jumlah total pengguna sepeda baik casual maupun registered?

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(
    x='temp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Temperature vs Count')

plt.subplot(1, 3, 2)
sns.scatterplot(
    x='atemp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Feels Like Temperature vs Count')

plt.subplot(1, 3, 3)
sns.scatterplot(
    x='hum',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Humidity vs Count')
```

Dari hasil analisis diatas, beberapa pernyataan yang dapat dibuat yaitu:
- Variabel temp dan count menunjukan korelasi yang positif
- Variabel atemp dan count juga menunjukan korelasi yang positif
- Variabel hum dan count menunjukan korelasi yang sedikit negatif, namun tidak terlalu signifikan

Kesimpulannya yaitu jika temp dan atemp meningkat, penggunaan sepeda juga cenderung meningkat. Dan pengaruh dari humidity sangatlah kecil

### Pertanyaan 3: Berapa banyak jumlah sepeda yang digunakan pada workingday, holiday, dan weekday?

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(15,10))

sns.barplot(
    x='workingday',
    y='count',
    data=day_df,
    ax=axes[0],
    palette='Set2')
axes[0].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Kerja (Workingday)')
axes[0].set_xlabel('Hari Kerja')
axes[0].set_ylabel('Jumlah Pengguna Sepeda')

sns.barplot(
  x='holiday',
  y='count',
  data=day_df,
  ax=axes[1],
  palette='Set2')
axes[1].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Libur (Holiday)')
axes[1].set_xlabel('Hari Libur')
axes[1].set_ylabel('Jumlah Pengguna Sepeda')

sns.barplot(
  x='weekday',
  y='count',
  data=day_df,
  ax=axes[2],
  palette='Set2')
axes[2].set_title('Jumlah Pengguna Sepeda berdasarkan Hari dalam Seminggu (Weekday)')
axes[2].set_xlabel('Hari dalam Seminggu')
axes[2].set_ylabel('Jumlah Pengguna Sepeda')

plt.tight_layout()
plt.show()
```

Dari hasil diatas, dapat disimpulkan bahwa:
- Jumlah penyewa sepeda lebih banyak pada hari kerja dibandingkan akhir pekan
- Jumlah penyewa sepeda lebih banyak pada hari biasa dibandingkan hari libur
- Jumlah penyewa sepeda pada hari Jumat adalah yang tertinggi, dan yang terendah ada pada hari Minggu
